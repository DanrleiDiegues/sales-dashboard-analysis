## Preprocess and Feature Engineering:

### Handling Missing Values

```python
# Handling missing values:
data["Segment"] = data["Segment"].fillna("Unknow")
data["Lead Source"] = data["Lead Source"].fillna("Unknow")
data["Type"] = data["Type"].fillna("Unknow")
data["Primary Campaign Source"] = data["Primary Campaign Source"].fillna("Unknow")
data["Close Reason"] = data["Close Reason"].fillna("Unknow")
data["Business Type"] = data["Business Type"].fillna("Unknow")
```

### # Update the numeric values to float and the problematic to NaN:

```python
# Update the numeric values to float and the problematic to NaN:
data[" Opp Value (EUR)"] = pd.to_numeric(data[" Opp Value (EUR)"], errors='coerce')
print(data[" Opp Value (EUR)"].dtype)
data[" Opp Value (EUR)"]
```

#### Creating a Deal Opened (Days) feature with Datatime:

```python
# "Close Date" and está no formato datetime

data[['Created Date', 'Close Date']]
```

```python
# Convert Created Date and Close Date to datetime and Garantee that the errors will be corrected:
data['Created Date'] = pd.to_datetime(data['Created Date'], format='%d/%m/%Y', errors="coerce")
data['Close Date'] = pd.to_datetime(data['Close Date'], format='%m/%d/%Y', errors="coerce")

# Calculate the time to close in days
data['Deal Opened (Days)'] = (data['Close Date'] - data['Created Date']).dt.days
data
```

```python
data_value_negative = data[data["Deal Opened (Days)"]<0]
data_value_negative
```

```python
# Remove rows where "Deal Opened (Days)" is negative
data = data[data["Deal Opened (Days)"] >= 0]
data
```
